```python

import matplotlib.pyplot as plt
import warnings
import random as ran
warnings.filterwarnings("ignore", category=UserWarning) # to ignore annoying warning messages when using matplotlib

import networkx as nx
```

```python
#Code for BA and ER link removal random
bal = nx.barabasi_albert_graph(2000,2)
counter = 0
bae = bal.number_of_edges()
ban =bal.number_of_nodes()
erl = nx.gnm_random_graph(ban, bae)
kb = bal.number_of_edges()/bal.number_of_nodes()
ke = erl.number_of_edges()/erl.number_of_nodes()
bagc = max(nx.connected_components(bal), key=len)
ergc = max(nx.connected_components(erl), key=len)
plotbak = []
ploterk = []
plotbagc = []
plotergc = []
plotbak.append(kb)
ploterk.append(ke)
plotbagc.append(len(bagc))
plotergc.append(len(ergc))

while (counter<bae-1):
    balinks = bal.edges()
    erlinks = erl.edges()
    toRemBA = ran.choice(balinks)
    toRemER = ran.choice(erlinks)
    bal.remove_edge(toRemBA[0],toRemBA[1])
    erl.remove_edge(toRemER[0],toRemER[1])
    counter+= 1
    kb = bal.number_of_edges()/bal.number_of_nodes()
    ke = erl.number_of_edges()/erl.number_of_nodes()
    bagc = max(nx.connected_components(bal), key=len)
    ergc = max(nx.connected_components(erl), key=len)
    plotbak.append(kb)
    ploterk.append(ke)
    plotbagc.append(len(bagc))
    plotergc.append(len(ergc))
#gc = max(nx.connected_component_subgraphs(bal), key=len)
print("done")
plt.plot(plotbak, 'ro')
plt.plot(ploterk, 'bo')
plt.show()
plt.plot(plotbagc, 'ro')
plt.plot(plotergc, 'bo')
plt.show()
```

```python
#Code for BA and ER node removal random
bag = nx.barabasi_albert_graph(2000,2)
counter = 0
baen = bag.number_of_edges()
bann =bag.number_of_nodes()
erg = nx.gnm_random_graph(bann, baen)
kb = bag.number_of_edges()/bag.number_of_nodes()
ke = erg.number_of_edges()/erg.number_of_nodes()
bagc = max(nx.connected_components(bag), key=len)
ergc = max(nx.connected_components(erg), key=len)
plotbak1 = []
ploterk1 = []
plotbagc1 = []
plotergc1 = []
plotbak1.append(kb)
ploterk1.append(ke)
plotbagc1.append(len(bagc))
plotergc1.append(len(ergc))
while (counter<ban-1):
    banodes = bag.nodes()
    ernodes = erg.nodes()
    toRemBA = ran.choice(banodes)
    toRemER = ran.choice(ernodes)
    bag.remove_node(toRemBA)
    erg.remove_node(toRemER)
    counter+= 1
    kb = bag.number_of_edges()/bag.number_of_nodes()
    ke = erg.number_of_edges()/erg.number_of_nodes()
    bagc = max(nx.connected_components(bag), key=len)
    ergc = max(nx.connected_components(erg), key=len)
    plotbak1.append(kb)
    ploterk1.append(ke)
    plotbagc1.append(len(bagc))
    plotergc1.append(len(ergc))
#gc = max(nx.connected_component_subgraphs(bal), key=len)
print("done")

plt.plot(plotbak1, 'ro')
plt.plot(ploterk1, 'bo')
plt.show()
plt.plot(plotbagc1, 'ro')
plt.plot(plotergc1, 'bo')
plt.show()
```
